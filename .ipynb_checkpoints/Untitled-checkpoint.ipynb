```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Convert to DataFrames for easier analysis
current_quiz_df = pd.DataFrame(current_quiz_data)
quiz_submission_df = pd.DataFrame(quiz_submission_data)
historical_data_df = pd.DataFrame(historical_data)

# Step 1: Analyze the Data
# Explore the schemas and inspect data
print("Current Quiz Data Sample:")
print(current_quiz_df.head())

print("\nQuiz Submission Data Sample:")
print(quiz_submission_df.head())

print("\nHistorical Data Sample:")
print(historical_data_df.head())

# Merge Data for Insights Analysis
merged_df = historical_data_df.copy()
if 'question_id' in quiz_submission_df.columns:
    merged_df = pd.merge(merged_df, quiz_submission_df, on='question_id', how='left')

# Calculate Topic-Wise Accuracy
def calculate_topic_accuracy(df):
    topic_accuracy = df.groupby('topic')['is_correct'].mean() * 100
    return topic_accuracy

def calculate_difficulty_accuracy(df):
    difficulty_accuracy = df.groupby('difficulty')['is_correct'].mean() * 100
    return difficulty_accuracy

topic_accuracy = calculate_topic_accuracy(merged_df)
difficulty_accuracy = calculate_difficulty_accuracy(merged_df)

print("\nTopic-Wise Accuracy:")
print(topic_accuracy)

print("\nDifficulty-Level Accuracy:")
print(difficulty_accuracy)

# Step 2: Generate Insights and Visualizations
sns.set(style="whitegrid")

# Plot topic-wise accuracy
plt.figure(figsize=(10, 6))
sns.barplot(x=topic_accuracy.index, y=topic_accuracy.values, palette="viridis")
plt.title("Topic-Wise Accuracy")
plt.xlabel("Topic")
plt.ylabel("Accuracy (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("topic_wise_accuracy.png")

# Plot difficulty-level accuracy
plt.figure(figsize=(8, 5))
sns.barplot(x=difficulty_accuracy.index, y=difficulty_accuracy.values, palette="coolwarm")
plt.title("Difficulty-Level Accuracy")
plt.xlabel("Difficulty Level")
plt.ylabel("Accuracy (%)")
plt.tight_layout()
plt.savefig("difficulty_level_accuracy.png")

# Step 3: Create Recommendations
def generate_recommendations(topic_accuracy, difficulty_accuracy):
    recommendations = []
    
    # Focus on weak topics
    weak_topics = topic_accuracy[topic_accuracy < 50].index.tolist()
    if weak_topics:
        recommendations.append(f"Focus on these weak topics: {', '.join(weak_topics)}")

    # Suggest improvement in difficulty levels
    weak_difficulties = difficulty_accuracy[difficulty_accuracy < 50].index.tolist()
    if weak_difficulties:
        recommendations.append(f"Practice questions of these difficulty levels: {', '.join(weak_difficulties)}")

    return recommendations

recommendations = generate_recommendations(topic_accuracy, difficulty_accuracy)
print("\nRecommendations:")
for rec in recommendations:
    print(f"- {rec}")

# Save recommendations to a file
with open("recommendations.txt", "w") as file:
    file.write("\n".join(recommendations))

print("\nRecommendations saved to recommendations.txt")
```
