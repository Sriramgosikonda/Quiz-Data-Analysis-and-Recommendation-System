# **Quiz Data Analysis and Recommendation System**

## **1. Project Overview**

- **Objective**: Analyze quiz data, generate insights, identify performance gaps, and provide personalized recommendations to users.
- **Scope**: Process and analyze both current and historical quiz data to deliver actionable insights and improvement suggestions.

---

## **2. Project Flow**

### **2.1 Data Understanding**

#### **2.1.1 Current Quiz Data**
- Extract details of the user's latest quiz submission.
  - Schema includes: `questions`, `topics`, `responses`, etc.

#### **2.1.2 Historical Quiz Data**
- Extract performance data for the last 5 quizzes of each user.
  - Schema includes: `scores`, `response map`, `difficulty levels`, etc.

---

### **2.2 Data Preprocessing**

#### **2.2.1 Cleaning Data**
- Handle missing or inconsistent values in both datasets.
- Normalize topic names and map question IDs with response IDs.

#### **2.2.2 Transformation**
- Extract:
  - Topic-wise performance
  - Difficulty levels
  - Response accuracy

#### **2.2.3 Data Merging**
- Combine current and historical data for deeper analysis.

---

### **2.3 Data Analysis**

#### **2.3.1 User Performance Analysis**
- Identify:
  - **Top-performing topics** based on accuracy and score.
  - **Weak areas**: Topics/questions with consistently low scores.
  - **Performance by difficulty level**.

#### **2.3.2 Trends Analysis**
- Track performance improvements or declines across the last 5 quizzes.
- Correlate response patterns with question difficulty.

#### **2.3.3 Response Accuracy Analysis**
- Evaluate:
  - Correct vs incorrect responses.
  - Errors by difficulty level.

---

### **2.4 Insights and Recommendations**

#### **2.4.1 Insights Generation**
- Highlight **strengths** and **weaknesses**.
  - Example: "User excels in algebra but struggles with geometry."
- Identify **specific performance gaps**:
  - Example: "User has difficulty with medium-difficulty problems in topic X."

#### **2.4.2 Recommendations**
- Propose actionable steps:
  - Focus on weaker topics.
  - Practice questions of specific difficulty levels.
  - Take additional quizzes with similar structures.

---

### **2.5 Student Persona Analysis (Bonus)**
- Define a detailed persona based on patterns.
  - **Strengths**: High performance in specific topics or question types.
  - **Weaknesses**: Areas needing focus.
  - **Learning Style**: Adaptability with question difficulty, response speed, etc.

---

## **3. Output Generation**

### **3.1 Visualizations**
- Topic-wise accuracy and scores (Bar Charts).
- Trends over 5 quizzes (Line Charts).
- Distribution of response accuracy (Pie Charts).

### **3.2 Summary Insights**
- Top 3 strengths and weaknesses.
- Progress trend analysis.

### **3.3 Personalized Recommendations**
- Actionable steps summarized as:
  - Topics to prioritize.
  - Suggested question types.
  - Time management tips.

```python
def convert_to_dataframe(data):
    if isinstance(data, list):
        # Directly create DataFrame for lists
        return pd.DataFrame(data)
    elif isinstance(data, dict):
        # Normalize dictionary (useful if it has nested structures)
        try:
            return pd.json_normalize(data)
        except Exception as e:
            print("Error normalizing dictionary:", e)
            return pd.DataFrame([data])  # Fallback to wrapping as a single-row DataFrame
    else:
        print(f"Unexpected data format: {type(data)}")
        return None

# Apply the function to datasets
current_quiz_df = convert_to_dataframe(current_quiz_data)
quiz_submission_df = convert_to_dataframe(quiz_submission_data)
historical_data_df = convert_to_dataframe(historical_data)

# Display the resulting DataFrames
print("Current Quiz DataFrame:")
print(current_quiz_df.head())

print("\nQuiz Submission DataFrame:")
print(quiz_submission_df.head())

print("\nHistorical Data DataFrame:")
print(historical_data_df.head())
```

```python
weak_topics = topic_performance[topic_performance['accuracy'] < 50]
print("\nWeak Areas by Topic:")
print(weak_topics)
```

```python
# Grouping by the number of questions in the quiz
difficulty_performance = historical_data_df.groupby('total_questions')['score'].mean().reset_index()
print("\nDifficulty Performance (based on questions count):")
print(difficulty_performance)
```

```python
# Assuming topic_performance contains the performance data for different topics
strengths = topic_performance[topic_performance['accuracy'] >= 80]['quiz_topic'].tolist()

# Assuming weak_topics contains topics with lower accuracy
weaknesses = weak_topics['quiz_topic'].tolist()

# Define the student's persona based on strengths and weaknesses
persona = {
    "Strengths": strengths,
    "Weaknesses": weaknesses,
    "Learning Style": "Prefers mastering specific topics and requires additional practice in challenging areas to increase overall performance."
}

# Print the student persona
print("\nStudent Persona:")
print(persona)
```

```python
import pandas as pd

# Identify weak areas based on accuracy less than a certain threshold (e.g., 60%)
weak_areas = topic_performance[topic_performance['accuracy'] < 60]

# Highlight the topics where performance gaps exist (based on weak accuracy)
print("\nWeak Areas (Accuracy < 60%):")
print(weak_areas)

# Track improvement trends (user score over time)
performance_trends = historical_data_df.groupby(['user_id', 'submitted_at'])['score'].mean().reset_index()
performance_trends['submitted_at'] = pd.to_datetime(performance_trends['submitted_at'])  # Ensure datetime format
performance_trends.set_index('submitted_at', inplace=True)
performance_trends = performance_trends.sort_index()

# Display trends for the given user
print("\nImprovement Trends:")
print(performance_trends)


# Analyze performance gaps by looking at initial mistakes (missed questions)
performance_gaps = historical_data_df.groupby('user_id')['initial_mistake_count'].sum().reset_index()
performance_gaps['user_id'] = performance_gaps['user_id'].astype(str)

print("\nPerformance Gaps (Mistakes):")
print(performance_gaps)

# Combine insights into a recommendation or action plan
recommendations = {
    "Weak Areas": weak_areas['quiz_topic'].tolist(),
    "Improve Accuracy by Revisiting": weak_areas['quiz_topic'].tolist(),
    "Work on Consistency in Mistakes": "Review missed questions and repeated mistakes over time.",
}

print("\nRecommendations for Improvement:")
print(recommendations)
```
